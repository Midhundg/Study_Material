# Permutation

```python
def permute(l):
    res=[]
    def backtrack(path,used):
        if len(path)==len(l):
            res.append(path[:])
            return
        for i in range(len(l)):
            if used[i]:
                continue
            used[i]=True
            path.append(l[i])
            backtrack(path,used)
            path.pop()
            used[i]=False
    backtrack([],[False]*len(l))
    return res
print(permute(['a','b','c']))
```
