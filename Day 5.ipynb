# Longest Balanced Subarray 1

```python
def longestBalanced(nums):
        n = len(nums)
        ans = 0
        for i in range(n):
            even_set = set()
            odd_set = set()
            for j in range(i, n):
                if nums[j] % 2 == 0:
                    even_set.add(nums[j])
                else:
                    odd_set.add(nums[j])
                if len(even_set) == len(odd_set):
                    ans = max(ans, j - i + 1)
        return ans
print(longestBalanced([3,2,2,5,4]))
```

# Peak Element 1

```python
def solve(arr,left,right):
    if left==right:
        return arr[left]
    mid=(left+right)//2
    if (mid==0 or arr[mid]>=arr[mid-1]) and (mid==len(arr)-1 or arr[mid]>=arr[mid+1]):
        return arr[mid]
    if mid>0 and arr[mid-1]>arr[mid]:
        return solve(arr,left,mid-1)
    else:
        return solve(arr,mid+1,right)
print(solve([3,2,1,5,6,4],0,5))
    
```

# Peak Element 2

```python
def solve(arr,left,right):
    if left>right:
        return float('-inf')
    if left==right:
        return arr[left]
    mid=(left+right)//2
    best=float('-inf')
    if (mid==0 or arr[mid]>=arr[mid-1]) and (mid==len(arr)-1 or arr[mid]>=arr[mid+1]):
        best=arr[mid]
    left_best=solve(arr,left,mid-1)
    right_best=solve(arr,mid+1,right)
    if left_best>best:
        best=left_best
    if right_best>best:
        best=right_best
    return best
print(solve([3,2,1,5,6,4],0,5))
    
```
