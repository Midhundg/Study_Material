# Partition Equal Subset Sum

```python
def canPartition(nums):
    if len(nums) == 1:
        return False
    total = sum(nums)
    if total % 2 != 0:
        return False
    k = total // 2
    def subset(nums, k):
        n = len(nums)
        dp = [[False] * (k + 1) for _ in range(n + 1)]
        for i in range(n + 1):
            dp[i][0] = True
        for i in range(1, n + 1):
            for s in range(1, k + 1):
                nopick = dp[i - 1][s]
                pick = False
                if nums[i - 1] <= s:
                    pick = dp[i - 1][s - nums[i - 1]]
                dp[i][s] = nopick or pick
        return dp[n][k]
    return subset(nums, k)
nums = [1, 5, 11, 5]
print(canPartition(nums))  # True
```

# Jump Game

```python
def canJump(nums):
        farthest=0
        for i in range(len(nums)):
            if i>farthest:
                return False
            farthest=max(farthest,i+nums[i])
        return True
print(canJump([2,3,1,1,4]))
```

# Coin Change

```python
def coinChange(coins,amount):
    n=len(coins)
    inf=float('inf')
    dp=[[inf]*(amount+1) for i in range(n+1)]
    for i in range(n+1):
        dp[i][0]=0
    for i in range(1,n+1):
        for j in range(1,amount+1):
            if coins[i-1]<=j:
                dp[i][j]=min(dp[i-1][j],1+dp[i][j-coins[i-1]])
            else:
                dp[i][j]=dp[i-1][j]
    return -1 if dp[n][amount]==inf else dp[n][amount]
print(coinChange([1,2,5],11))
```
