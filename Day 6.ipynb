# Partition Equal Subset Sum

```python
def partition(nums):
    s=sum(nums)
    if s%2!=0:
        return False
    total=s//2
    n=len(nums)
    dp=[[False]*(total+1) for i in range(n+1)]
    for i in range(n+1):
        dp[i][0]=True
    for i in range(1,n+1):
        for j in range(1,total+1):
            if nums[i-1]<=j:
                dp[i][j]=dp[i-1][j] or dp[i-1][j-nums[i-1]]
            else:
                dp[i][j]=dp[i-1][j]
    return dp[n][total]
print(partition([1,5,11,5]))
    
```

# Jump Game

```python
def canJump(nums):
        farthest=0
        for i in range(len(nums)):
            if i>farthest:
                return False
            farthest=max(farthest,i+nums[i])
        return True
print(canJump([2,3,1,1,4]))
```

# Coin Change

```python
def coinChange(coins,amount):
    n=len(coins)
    inf=float('inf')
    dp=[[inf]*(amount+1) for i in range(n+1)]
    for i in range(n+1):
        dp[i][0]=0
    for i in range(1,n+1):
        for j in range(1,amount+1):
            if coins[i-1]<=j:
                dp[i][j]=min(dp[i-1][j],1+dp[i][j-coins[i-1]])
            else:
                dp[i][j]=dp[i-1][j]
    return -1 if dp[n][amount]==inf else dp[n][amount]
print(coinChange([1,2,5],11))
```

# Longest Common Subsequence


```python
def LCS(s1,s2):
    m=len(s1)
    n=len(s2)
    dp=[[0]*(m+1) for i in range(n+1)]
    for i in range(1,n+1):
        for j in range(1,m+1):
            if s1[j-1]==s2[i-1]:
                dp[i][j]=dp[i-1][j-1]+1
            else:
                dp[i][j]=max(dp[i-1][j],dp[i][j-1])
    return dp[n][m]
print(LCS('abcde','ace'))
```

# Divisor Game

```python
def divisorGame(n):
        dp=[False]*(n+1)
        dp[1]=False
        for i in range(2,n+1):
            for x in range(1,i):
                if i%x==0 and dp[i-x]==False:
                    dp[i]=True
                    break
        return dp[n]
print(divisorGame(3))
```
