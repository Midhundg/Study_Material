# Pascal's Triangle

```python
def Pascals(nums):
    if nums==0:
        return []
    dp=[]
    for i in range(nums):
        res=[1]*(i+1)
        for j in range(1,i):
            res[j]=dp[i-1][j-1]+dp[i-1][j]
        dp.append(res)
    return dp
print(Pascals(5))
```

# Pascal's Triangle 2

```python
def getRow(rowIndex):
    dp=[]
    for i in range(rowIndex+1):
        res=[1]*(i+1)
        for j in range(1,i):
            res[j]=dp[i-1][j-1]+dp[i-1][j]
        dp.append(res)
    return dp[rowIndex]
print(getRow(3))
```

# Flood Fill

```python
def floodFill(image,sr,sc,color):
        n=len(image)
        m=len(image[0])
        start=image[sr][sc]
        def bt(x,y):
            if start==color:
                return image
            if x<0 or y<0 or x>=n or y>=m or image[x][y]!=start:
                return False
            image[x][y]=color
            if bt(x,y+1):
                return True
            if bt(x+1,y):
                return True
            if bt(x-1,y):
                return True
            if bt(x,y-1):
                return True
        bt(sr,sc)
        return image
```
