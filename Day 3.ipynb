# Pascal's Triangle

```python
def Pascals(nums):
    dp=[]
    for i in range(nums):
        rows=[1]*(i+1)
        for j in range(1,i):
            rows[j]=dp[i-1][j-1]+dp[i-1][j]
        dp.append(rows)
    return dp
print(Pascals(5))
```

# Pascal's Triangle 2

```python
def getRow(rowIndex):
        if rowIndex>0:
            dp=[]
            for i in range(rowIndex+1):
                row=[1]*(i+1)
                for j in range(1,i):
                    row[j]=dp[i-1][j-1]+dp[i-1][j]
                dp.append(row)
            return dp[rowIndex]
        else:
            return [1]
print(getRow(3))
```

# Flood Fill

```python
def floodFill(image,sr,sc,color):
        n=len(image)
        m=len(image[0])
        start=image[sr][sc]
        def bt(x,y):
            if start==color:
                return image
            if x<0 or y<0 or x>=n or y>=m or image[x][y]!=start:
                return False
            image[x][y]=color
            if bt(x,y+1):
                return True
            if bt(x+1,y):
                return True
            if bt(x-1,y):
                return True
            if bt(x,y-1):
                return True
        bt(sr,sc)
        return image
```
