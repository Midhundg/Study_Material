# Combination Sum

```python
nums = [1,2,3]
n = len(nums)
path = []
res = []
def solve(start):
    path.append(res[:])
    if start>n:
        return 
    for i in range(start,n):
        res.append(nums[i])
        solve(i+1)
        res.pop()
solve(0)
print(path)
```

# Subset (Leetcode 78)

```python
nums=[1,2,3]
n=len(nums)
res=[]
path=[]
def solve(start):
    path.append(res[:])
    if start>n:
        return
    for i in range(start,n):
        res.append(nums[i])
        solve(i+1)
        res.pop()
solve(0)
print(path)
```
