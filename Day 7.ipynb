# Two City Scheduling


```python
def twoCitySchedCost(costs):
        costs.sort(key=lambda x:x[0]-x[1])
        n=len(costs)//2
        total=0
        for i in range(len(costs)):
            if i<n:
                total+=costs[i][0]
            else:
                total+=costs[i][1]
        return total
print(twoCitySchedCost([[10,20],[30,200],[400,50],[30,20]]))
```

# Climbing Stairs

```python
def climbStairs(n):
        if n==1:
            return n
        if n==2:
            return n
        dp=[0]*(n+1)
        dp[1]=1
        dp[2]=2
        for i in range(3,n+1):
            dp[i]=dp[i-1]+dp[i-2]
        return dp[n]
print(climbStairs(5))
```

# Majority Element

```python
def majorityElement(nums):
        if len(nums)==1:
            return nums[0]

        mid=len(nums)//2

        left_ans=majorityElement(nums[:mid])
        right_ans=majorityElement(nums[mid:])

        if left_ans==right_ans:
            if nums.count(left_ans)>len(nums)//2:
                return left_ans
            else:
                return -1

        left_ans1=nums.count(left_ans)
        right_ans1=nums.count(right_ans)

        if left_ans1>len(nums)//2:
            return left_ans
        if right_ans1>len(nums)//2:
            return right_ans
        return -1
print(majorityElement([3,2,3]))

        
```

# Leetcode 135.Candy


```python
def candy(ratings):
        n=len(ratings)
        if n==0:
            return 0
        candies=[1]*n
        for i in range(1,n):
            if ratings[i]>ratings[i-1]:
                candies[i]=candies[i-1]+1
        for i in range(n-2,-1,-1):
            if ratings[i]>ratings[i+1]:
                candies[i]=max(candies[i],candies[i+1]+1)
        return sum(candies)
print(candy([1,0,2]))
```

# 198. House Robber

```python
def house_robber(nums):
    n=len(nums)
    if n==0:
        return 0
    if n==1:
        return nums[0]
    dp=[0]*n
    dp[0]=nums[0]
    dp[1]=max(dp[0],dp[1])
    for i in range(2,n):
        dp[i]=max(dp[i-1],nums[i]+dp[i-2])
    return dp[n-1]
print(house_robber([1,2,3,1]))
```

# 105. Construct Binary Tree from Preorder and Inorder Traversal

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def buildTree(self, preorder, inorder):
        if not preorder or not inorder:
            return None

        root = TreeNode(preorder[0])
        mid = inorder.index(preorder[0])

        root.left = self.buildTree(preorder[1:mid+1], inorder[:mid])
        root.right = self.buildTree(preorder[mid+1:], inorder[mid+1:])

        return root

s = Solution()
root = s.buildTree([3,9,20,15,7], [9,3,15,20,7])
print(root.val)          # prints 3
print(root.left.val)     # prints 9
print(root.right.val)    # prints 20
```
